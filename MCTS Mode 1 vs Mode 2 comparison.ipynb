# MCTS Mode 1 vs Mode 2 comparison

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from numpy.linalg import pinv, norm
seaborn.set()
%matplotlib inline  
```

## First scenario: One pedestrian at the edge of the road

```python
scen110_rewards = pd.read_table("mode2/scen1.1/scen110_rewards.txt", header = None)
scen11_rewards = pd.read_table("mode2/scen1.1/scen11_rewards.txt", header = None)
```

```python
plt.rc("figure", figsize=(22,6))
plt.plot(scen110_rewards, label = "mode1")
plt.plot(scen11_rewards, label = "mode2")
plt.xlabel("Calls to step function / 100")
plt.ylabel("Reward")
plt.legend()
```

The jump in the plot is when an event is found. The first mode has less calls to the step function because it teminates once it commits to a node that has an event.

## Comments:
- The first mode generally finds the optimal path at a lower number of calls, but continues to run until it commits to the event node, while the second mode finds the optimal path later, but it can terminate there.
- The multiple pedestrian scenarios did not show any difference than a single pedestrian.
- All of these results are from one seed only, and performed to gain some insights into the differences between the two modes and the six scenarios. More runs with more seeds are needed to have more comprehensive conclusion.
